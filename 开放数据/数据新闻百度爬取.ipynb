```python
import pandas as pd

zisha_data = pd.read_excel("/Users/Yufeng1/Desktop/demo/BaiduSpider/result.xlsx")
print(zisha_data.head())
```

```python
des = zisha_data['描述']
print(des.head())
```

```python
import jieba

sentence_seged = []
fenci = []
for i in des:
    seged = []
    sentence_seged = jieba.cut(i)
    for j in sentence_seged:
        if len(j) > 1:
            seged.append(j)
    fenci.append(seged)

print(fenci[2])
print(len(des), len(fenci))
```

```python
from collections import Counter

words = []
for i in fenci:
    for j in i:
        words.append(j)
Counter(words)
```
